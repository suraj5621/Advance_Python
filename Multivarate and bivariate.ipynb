```python
import pandas as pd
import seaborn as sns

tips=sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
print(tips.sample(5))
print('------------------------------------------------------')
print('------------------------------------------------------')
print(titanic.sample(5))
print('------------------------------------------------------')
print('------------------------------------------------------')
print(flights.sample(5))
print('------------------------------------------------------')
print('------------------------------------------------------')
print(iris.sample(5))
print('------------------------------------------------------')
print('------------------------------------------------------')
```

## Scatter plot (Numerical  ~ Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'] , hue=tips['sex'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'] , hue=tips['sex'] , style=tips['time'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'] , hue=tips['sex'] , style=tips['time'] , size=tips['day'] )
```

## BoxPlot (Numerical ~  Categorical)

```python
sns.boxplot(x=titanic['pclass'], y=titanic['age'])
```

```python
sns.boxplot(x=titanic['pclass'], y=titanic['age'], hue=titanic['sex'])
```

## Distplot

```python
sns.distplot(titanic[titanic['survived'] == 0]['age'] , hist=False)
sns.distplot(titanic[titanic['survived'] == 1]['age'] , hist=False)
```
