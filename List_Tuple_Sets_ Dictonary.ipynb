```python
# Repetition
print('Repetition')
list1=[2,3,4,5,6]
print(list1*2)
print('-------------------------------------')
# Concatenation
print('Concatenation')
list2=[7,3,9,1]
print(list1+list2)
print('-------------------------------------')
# Length
print('length')
print(len(list1))
print('-------------------------------------')
# iteration
print('iteration')
for i in list1:
  print(i)
print('-------------------------------------')
# membership
print('membership')
a=7 in list2
print(a)
print('-------------------------------------')
```

# Sets
* A set is a collection which is unordered, unchangeable*, and unindexed.
* Duplicates Not Allowed
* Although  sets are unchangeable but we can update or remove element

```python
set1 = {1,2,3, "JavaTpoint", 20.5, 14}  
print(set1)
print(type(set1))

```

```python
# add element
set1.add('python')
print(set1)
```


* clear() -->    Removes all the elements from the set 
* pop()   -->   	Removes an element from the set
* remove() -->    Removes the specified element
* discard() ->    Remove the specified item
** remove and discard are similar

```python
# pop
print('pop')
set1.pop()
print(set1)
print('----------------------------------------')
# remove
print('remove')
set1.remove(2)
print(set1)
print('----------------------------------------')
# discard
print('discard')
set1.discard('python')
print(set1)
print('----------------------------------------')
# clear
print('clear')
set1.clear()
print(set1)
```

-------------------------------------------------------
* intersection
* union

```python
set1={2,8,5,3,9}
set2={8,6,1,2}
set3=set1.intersection(set2)
print('intersection')
print(set3)
set4=set1.union(set2)
print('union')
print(set4)
```

# Dictionary
* Dictionaries are used to store data values in key:value pairs.
* A dictionary is a collection which is ordered*, changeable
* As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.
* Dictionaries are written with curly brackets, and have keys and values
* All key are always unique and Keys must only have one component. 
* values can be duplicates and Values can be of any type, including integer, list, and tuple.

```python
dict1 = {1: 'Hcl', 2: 'WIPRO', 3:'Facebook'}
print(dict1) 
print(type(dict1))
print(len(dict1))

print('for all keys we use dict1.keys() --> ' , dict1.keys())
print('for all keys we use dict1.values() --> ' , dict1.values())
```

```python
# Accessing
print(dict1[2])

dict2={"brand": "Ford", "model": "Mustang", "year": 1964}
print(dict2['brand'])
print('----------------------------------------')


# Accessing by using get keyword
print('Accessing by using get keyword')
print(dict2.get('model'))
print('----------------------------------------')

# by using item() method
print('by using item() method')
print(dict2.items())
```

### Adding

```python
# Adding elements to dictionary one at a time       
dict2[0] = 'Peter'      

# Adding multiple elements to dictionary one at a time   
dict2['Emp_ages'] = 20, 33, 24     

# Updating existing Key's Value       
dict2['year'] = 1924      


# Updating existing Key's Value by using update keyword       
dict2.update({'year': 1969})


print(dict2)
```

### remove

```python
# by using pop method
dict2.pop("model")

# by using popitem() method
# popitem() ---->   Remove the last item 
dict2.popitem()

print(dict2)
```
