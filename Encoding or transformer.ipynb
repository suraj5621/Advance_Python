```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

## Transformer without pipeline

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer= ColumnTransformer(transformers=[('trans_1' , SimpleImputer() ,['fever'] ) ,
                                             ('trans_2' , OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
                                             ('trans_3' , OneHotEncoder(sparse=False) , ['city' , 'gender'])
                                             ])
```

```python
X_train=transformer.fit(X_train)
```

```python
transformer.transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
