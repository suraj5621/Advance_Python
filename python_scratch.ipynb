##If Elif and Else

```python
number = int(input("Enter the number?"))  
if number==10:  
    print("number is equals to 10")  
elif number==50:  
    print("number is equal to 50");  
elif number==100:  
    print("number is equal to 100");  
else:  
    print("number is not equal to 10, 50 or 100");  
```

##Python Loops
* while --> when no. of iteration is unknown
* for   --> when no. of iteration is known

while <condition>:  
    { code block }  

--------------------------------------------------

for value in sequence:  
    { code block }  

```python
# Normal while loop
counter = 0
while counter < 10:
    counter = counter + 3  
    print("Python Loops")  
```

```python
# while loop with else
counter = 0    
while counter < 10: 
    counter = counter + 3  
    print("Python Loops")  
# Once the condition of while loop gives False this statement will be executed  
else:  
    print("Code block inside the else statement")  
```

```python
# Normal for loop
tuple_ = (3, 4, 6, 8, 9, 2, 3, 8, 9, 7)  
for i in tuple_: 
    print(i) 
```

```python
# for loop with else
string = "Python Loop"
for i in string: 
    if i == "o":  
        print("If block") 
    else:  
        print(i)  
else:
  print('loop is end')
```

```python
# for loop  with range
for i in range(5):
  print(i)
```

## Break , continue , pass
* Break statement	--> This command terminates the loop's execution and transfers the program's control to the statement next to the loop.

*	Continue statement -->	This command skips the current iteration of the loop. 

*	Pass statement -->	The pass statement is used when a statement is syntactically necessary, but no code is to be executed.
