* Min-max scaler
  * x`i = (xi - x min)/ (x_max - x_min)
  * isme minimum wala zero pr put krte hai or maximum wala 1 pr rkhte hai baki ko squeez krte hai
  * iska use tbb krte hai jbb aapko value ki limit pta ho jaise images ke mamale mein humko pta hai ki 0 to 255 ki range hoti hai to use krna better option hota hai

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

* mean scaling
  * x`i = (xi - x_mean)/(x_max - x_min)
  * range = -1 to 1
  * if xi < mean = -ve
  * else xi is +ve

* robust scaling
  * x`i = (xi -  median)/ IQR
  * uses when more outlier present

## Encoding Categorical data
* ordinal encoding ---> ordinal categorical data
* label encoding
* one hot encoding ---> nominal categorical data

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test)
```
