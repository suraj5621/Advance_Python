###Strings indexing and splitting

```python
str = "HELLO"  
# positive indexing
print('positive indexing')
print(str[0])  
print(str[1])  
print(str[2]) 
print('--------------------------')
# negative indexing
print('negative indexing')
print(str[-1])  
print(str[-2])  
print(str[-3]) 
```

###Slicing
str="Python_String"
str[ start : end : skip]

```python
str = "Python_String"  
print('slicing or splitting with positive index')
# Start Oth index to end  
print(str[0:])  
# Starts 1th index to 4th index  
print(str[1:5])
# Starts 0th to 2nd index  
print(str[:3]) 
print('-----------------------------------------------------')
# with skiping item
print('with skiping item')
print(str[2:5:2])
```

```python
print('slicing or splitting with negative index')
str = 'JAVATPOINT'  
print(str[-1])  
print(str[-3])  
print(str[-2:])  
print(str[-4:-1])  
print(str[-7:-2])  
# Reversing the given string
print('-----------------------------------------------------')
print('Reversing the given string')  
print(str[::-1]) 
```

#Functions
* User-defined and built-in functions are the two main categories of functions in Python
* Python programming language are provided by reference. not callby value


* Function Arguments
    * Default arguments
    * Keyword arguments
    * Required arguments
    * Variable-length arguments

```python
# defalt argument
def function( n1, n2 = 20 ):    
    print("number 1 is: ", n1)    
    print("number 2 is: ", n2)  

function(30)  
print('---------------------------------')
function(50,30)    
```

```python
# keyword argument

print( "Without using keyword" )    
function( 50, 30)  
print( "With using keyword" )    
function( n2 = 50, n1 = 30)    
```

```python
# Required Arguments
print( "Passing out of order arguments" )    
function( 30, 20 )       

print( "Passing only one argument" )    
try:    
    function( 30 )    
except:    
    print( "Function needs two positional arguments" )  
```
